## Analysis of Start_Time column

### Ques. What time of the day are accidents most frequent in? 

```python
df.Start_Time
```

```python
df.Start_Time[0]
```

```python
df.Start_Time =pd.to_datetime(df.Start_Time)
```

```python
df.Start_Time[0]
```

```python
df.Start_Time.dt.hour
```

```python
sns.distplot(df.Start_Time.dt.hour, bins=24, kde=False, norm_hist=True)
```

## Conclusions-
1. A high percentage of accidents occurs between 2pm to 5pm.
2. Next time range for most accidents are from 6am to 8am.
3. Number of Accidents starts to decrease from 5pm to 4am.


## Which days of the week have the most accidents?

```python
sns.distplot(df.Start_Time.dt.dayofweek, bins=7, kde=False)
```

### Conclusions -
1. Accidents are low on weekends (Saturday and Sunday)
2. All other days have almost same number of accidents.

###  QUES - Is the distribution of accidents by hour the same on weekends as on weekdays?

```python
sunday_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 6]
sns.distplot(sunday_start_time.dt.hour, bins=24, kde=False, norm_hist=True)
```

```python
saturday_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 5]
sns.distplot(saturday_start_time.dt.hour, bins=24, kde=False, norm_hist=True)
```

```python
monday_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 0]
sns.distplot(monday_start_time.dt.hour, bins=24, kde=False, norm_hist=True)
```

## Conclclusions -
1. On weekends (Sundays and Saturdays) occurance of accidents increases during night as compared to weekdays.
2. On weekends occurance of accidents is at peak during 12noon to 5pm.

## Which months have the most accidents?

```python
sns.distplot(df.Start_Time.dt.month, bins=12, kde=False)
```

## Conclusion -
1. Number of accidents tends to increase drastically during month of October, November and December
2. Maximum number of accidents are in the month of December.
3. Minimum number of accidents are in the month of July.

```python
df_2018 = df[df.Start_Time.dt.year == 2018]
sns.distplot(df_2018.Start_Time.dt.month, bins=12, kde=False)
```

### Conclusion-

There is slight variation in 2018 accidents variaton over 12 month of a year than that of overall 6 years of data variation over 12 months.

This can be because there is incomplete data from 2016 and 2021 so it could affect the variation.



## What is trend of accidents year over year?

```python
sns.distplot(df.Start_Time.dt.year, bins=6, kde=False)
```

## Conclusion
1. Number of Accidents has increased exponentially.
2. Number of Accidents in 2021 increased almost 2.5 times to that of 2020.
3. Number of Accidents in 2020 increased tripled to that of 2019.
